```python
import json
import pandas as pd
from pprint import pprint
import os
```

## Load meta data and prep

```python
src_path, file_type = "../data/syrian_home_germany-04-05-2016_2_28_09/", ".json"
 
file_paths = []  
for root, dirs, files in os.walk(src_path):
    file_paths.extend([os.path.join(root, f) for f in files if f.endswith(file_type)])
    
print 'Number of files is', len(file_paths)
```

```python
df = pd.read_json(file_paths[0])
print 'Data shape is', df.shape

df
```

## Load posts and compute stats

```python
keys_to_study = ['from', 'type', 'updated_time', 'created_time', 'picture']
```

```python
data_frames = []
for file_path in file_paths:
    df = pd.read_json(file_path)
    data_frames.append(df[keys_to_study])
```

```python
result = pd.concat(data_frames)
```

```python
result.head()
```

```python
result['id'] = result['from'].apply(lambda x: x['id'])
```

```python
result.to_csv("../data/syrian_home_germany-04-05-2016_2_28_09-SUMMARY.csv", index = False)
```

## Download images

```python
import urllib2
file_urls = list(result['picture'].dropna())
print(len(file_urls))
```

```python
dir_to_save = "../imgs/"
file_ids = []
for i, file_url in enumerate(file_urls):
    print i/float(len(file_urls))
    try:
        url = file_url
        filename = str(i) + ".jpg"
        file_ids.append({'url': url, 'filename': filename})
        img = urllib2.urlopen(url).read()
        output = open(os.path.join(dir_to_save,filename),'wb')
        output.write(img)
        output.close()
    except:
        print url, 'failed' 
```
